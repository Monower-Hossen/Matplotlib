## line Plot

```python
import matplotlib.pyplot as plt 

x=['Math','English','Bangla']
y = [75,77,80]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 

x=['Math','English','Bangla']
y = [75,77,80]

plt.plot(x,y)

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Graph', fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt 

sub=['Math','English','Bangla']
x1 = [75,77,80]
y1 =[50,88,85]

plt.plot(sub,x1,label='Shamim')
plt.plot(sub,y1,label='Tamanna')

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Graph', fontsize = 15)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 

sub=['Math','English','Bangla']
x1 = [75,77,80]
y1 =[50,88,85]

plt.plot(sub,x1,color='g', linewidth =1,linestyle = ':',marker='o',label='Shamim')
plt.plot(sub,y1,color ='r',marker= 'o', label='Tamanna')

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Graph', fontsize = 15)
plt.legend()
plt.show()
```
