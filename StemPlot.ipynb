# Stem Plot

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y=[2,6,5,7,5,1]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y=[2,6,5,7,5,1]
plt.stem(x,y,linefmt = ":", markerfmt = "ro")
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y=[2,6,5,7,5,1]
plt.stem(x,y,linefmt = ":", markerfmt = "ro",bottom = 9)
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y=[2,6,5,7,5,1]
plt.stem(x,y,linefmt = ":", markerfmt = "ro",basefmt="g")
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y=[2,6,5,7,5,1]
plt.stem(x,y,linefmt = ":", markerfmt = "ro",basefmt="g",label = "Python")
plt.legend()
plt.show()
```
