# Step Plot

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,2,5,4]

plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,2,5,4]

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.step(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,2,5,4]

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.step(x,y,color='r',marker ='o',ms=10)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,2,5,4]

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.step(x,y,color='r',marker ='o',ms=10,mfc='g')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4,5]
y= [2,3,2,5,4]

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.step(x,y,color='r',marker ='o',ms=10,mfc='g',label= "python")
plt.legend(loc= 3)
plt.grid()
plt.show()
```
