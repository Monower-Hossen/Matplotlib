# Axis Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,7]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,7]

plt.plot(x,y)

plt.xticks(x,xlabels=['python ', 'java','c','c++'])   
plt.yticks(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,7]

plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,10)

plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,7]

plt.plot(x,y)

plt.axis([0,10,0,7])
plt.show()
```
