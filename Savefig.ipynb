# Savefig 

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,4,6,4]
plt.plot(x,y,color = 'r')

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,4,6,4]
plt.plot(x,y,color = 'r')
plt.savefig('l1.pdf',dpi = 2000)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,4,6,4]
plt.plot(x,y,color = 'r')
plt.savefig('qq',dpi = 2000,facecolor = 'g',transparent = True)
plt.show()
```
