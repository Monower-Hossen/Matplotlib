## Bar Plot

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =20)

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =15)

c = ['y','b','g','m']

plt.bar(x,y,width = 0.6 , color = c,align = "center",edgecolor = "g",linewidth =5,linestyle = ":",alpha = 0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =15)

c = ['y','b','g','m']

plt.bar(x,y,width = 0.6 , color = c,align = "center",edgecolor = "r",label = "Popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]
z=[20,30,20,14]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =15)

c = ['y','b','g','m']
plt.bar(x,y,width = 0.4 , color = "r",label = "Popularity")
plt.bar(x,z,width = 0.4 , color = "y",label = "Popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]
z= [20,30,20,14]

width = 0.2
p= np.arange(len(x))
p1=[j+width for j  in p ]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =15)

plt.bar(p,y,width, color = "r",label = "Popularity")
plt.bar(p1,z,width,color = "y",label = "Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","C" ,"C++" ,"JAVA"]
y = [85,75,60,82]
z= [20,30,20,14]

plt.xlabel("language",fontsize =15)
plt.ylabel("No.",fontsize =15)
plt.title("Programming Language",fontsize =15)

plt.barh(p,y,width,color = "r",label = "Popularity")
plt.barh(p1,z,width,color = "y",label = "Popularity1")

plt.legend()
plt.show()
```
