## Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

plt.scatter(x,y)
plt.show()
```

## scatter fontsize

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

plt.scatter(x,y,color = 'r')

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```

## scatter color

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

colors = ['g','y','r','b','g','r','y']
plt.scatter(x,y, c=colors )

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```

## Size

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

colors = ['g','y','r','b','g','r','y']
sizes = [200,252,180,314,247,158,440]
plt.scatter(x,y, c=colors ,s=sizes , alpha =0.5,marker="*",edgecolor = "g",linewidth = 2)

plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

colors = [10,20,41,30,12,85,55]
sizes = [200,252,180,314,247,158,440]
plt.scatter(x,y, c=colors ,s=sizes ,cmap = "viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]

colors = [10,20,41,30,12,85,55]
sizes = [200,252,180,314,247,158,440]
plt.scatter(x,y, c=colors ,s=sizes ,cmap = "viridis")
p=plt.colorbar()
p.set_label("colorbar",fontsize =20)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y=[2,3,5,4,8,5,4]
z=[2,4,5,6,9,8,4]

colors = [10,20,41,30,12,85,55]
sizes = [200,252,180,314,247,158,440]

plt.scatter(x,y, c=colors ,s=sizes , cmap = "viridis")
plt.scatter(x,z, color='r' ,s=sizes, alpha =0.5 )

p=plt.colorbar()

p.set_label("colorbar",fontsize =20)
plt.title("Scatter Plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No", fontsize = 15)

plt.show()
```
